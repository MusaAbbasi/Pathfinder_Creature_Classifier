```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

### Pathfinder scrape

#### Traits main page

```python
#request the webpage
pathfinder_traits = requests.get("https://2e.aonprd.com/Traits.aspx")
```

```python
#turn it into soup
soup = BeautifulSoup(pathfinder_traits.text, "html.parser")
```

```python
#Save it so i dont have to request it a bunch
with open('pathfinder_traits.html', 'w', encoding='utf-8') as f:
    f.write(str(soup))
```

```python
with open('pathfinder_traits.html', 'r') as f:
    contents = f.read()
```

```python
contents
```

```python
soup = BeautifulSoup(contents, 'lxml')
```

```python
#Select only the part of the page i want
traits_section = soup.find("span", id="ctl00_RadDrawer1_Content_MainContent_DetailedOutput")
```

```python
#Trait name
trait_name = traits_section.findAll("span")[0].attrs['title']
trait_name
```

```python
trait_loc = traits_section.findAll("a")[0].attrs['href']
trait_loc
```

#### Each trait page

```python
trait_name = traits_section.findAll("span")[6].attrs['title']
trait_loc = traits_section.findAll("a")[6].attrs['href']
url = 'https://2e.aonprd.com/' + trait_loc
trait_1 = requests.get(url)
soup_1 = BeautifulSoup(trait_1.text, "html.parser")
x = soup_1.find("span", id="ctl00_RadDrawer1_Content_MainContent_DetailedOutput")
y = x.find_all(text=True)
z=[i for i in y if i!= " " and i!= "3.0"]
z
```

### Trait scraping details

```python
traits_dict = {}
name_list = []
value_list = []
```

```python

for n in range(0,477):
    trait_name = traits_section.findAll("span")[n].attrs['title']
    trait_loc = traits_section.findAll("a")[n].attrs['href']
    url = 'https://2e.aonprd.com/' + trait_loc
    trait_n = requests.get(url)
    soup_n = BeautifulSoup(trait_n.text, "html.parser")
    x = soup_n.find("span", id="ctl00_RadDrawer1_Content_MainContent_DetailedOutput")
    y = x.find_all(text=True)
    z=[i for i in y if i!= " " and i!= "3.0" and i!= "1.1" and i!= "2.0"]
    trait_value = z[3]
    name_list.append(trait_name)
    value_list.append(trait_value)
```

```python
print(name_list)
print(value_list)
```

```python
traits_df = pd.DataFrame()
```

```python
name_series = pd.Series(name_list)
value_series = pd.Series(value_list)
```

```python
traits_df["Name"] = name_series
traits_df["Description"] = value_series
traits_df.head()
```

```python
traits_df["Description"].value_counts()
```

```python
traits_df.to_csv("All_Traits.csv")
```
